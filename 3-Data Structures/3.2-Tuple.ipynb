Tuples


 tuples are orderd collections of items that are immutable. they are similar to list , but their immutability make them different.
 

```python
## creating a tuple

empty_tuple = ()
print("Empty Tuple:", empty_tuple)
print("Type of empty_tuple:", type(empty_tuple))
```

```python
lst=list()
print(type(lst))
tpl=tuple()
print(type(tpl))    
```

```python
numbers=tuple([1,2,3,4,5])
numbers
```

```python
list((1,2,3,4,5))
```

```python
mixed_tuple = (1, "Hello", 3.14, True)
print("Mixed Tuple:", mixed_tuple)
```

```python
## Accessing tuple elments

numbers
```

```python
print(numbers[2])
print(numbers[-1])
```

```python
numbers[0:3]
```

```python
numbers[1:]
```

```python
## reverse of tuples
numbers[::-1]
```

```python
## Tuple Operations

concatenation = numbers + mixed_tuple
print("Concatenated Tuple:", concatenation)
```

```python
mixed_tuple * 3
```

```python
numbers *3
```

```python
## immutable nature of tuples

lst=[1,2,3,4,5]
print(lst)

lst[1]="krish"
print(lst)
```

```python
## tuple Methods 

print(numbers.count(1))
print(numbers.index(3))
```

```python
## packing and unpacking tuples

#Packing
packed_tuple = 10, 20, 30
print("Packed Tuple:", packed_tuple)
```

```python
#Unpacking

a, b, c = packed_tuple
print("Unpacked Values:", a, b, c)
```

```python
## Unpacking with asterisk (*)

numbers=(1,2,3,4,5,6)
first, *middle, last = numbers
print("First:", first)
print("Middle:", middle)
print("Last:", last)
```

```python
## Nessted Tuples

#Nested list
lst=[[1,2,3],[4,5,6],[7,"hello",3.14]]
lst[0][0:2]

```

```python
#Nested Tuple
nested_tuple = ((1,2,3),(4,5,6),(7,"hello",3.14))
nested_tuple[0][0:3]
```

```python
## iterate over nested tuple

for sub_tuple in nested_tuple:
    for item in sub_tuple:
        print(item,end=" ")
    print()
```
