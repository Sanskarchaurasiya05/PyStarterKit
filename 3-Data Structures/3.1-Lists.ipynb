```python
## Acessing list elements

fruits = ["apple", "banana", "cherry", "date", "elderberry"]    
print(fruits[0])        # First element
print(fruits[2])        # Third element 
```

```python
print(fruits[-1])       # Last element
print(fruits[-3])       # Third from last element   
```

```python
print(fruits[1:])
print(fruits[:3])
print(fruits[1:4])
```

```python
print(fruits[-1:]) 
print(fruits[-1:-3])       # Last element
```

```python
## modifying list elements
fruits[1] = "blueberry"
print(fruits)
```

```python
fruits[1:]="watermelon"
## from one index everything is replaced by characters of watermelon
print(fruits)
```

```python
## List methods

fruits.append("orange") ## add an item to the end
print(fruits)
```

```python
fruits.insert(1,"kiwi") ## insert at a particular index
print(fruits)
```

```python
print(fruits)
fruits.remove("apple") ## it remove the first occurance of item
print(fruits)
```

```python
## remove and return last item
frt=fruits.pop()
print(frt)
print(fruits)
```

```python
#how to get index of an item
index_of_banana=fruits.index("kiwi")      
print(index_of_banana)
```

```python
fruits.insert(2,"kiwi") ## insert at a particular index
print(fruits.count("kiwi"))  ## count occurances of an item
```

```python
## sort entire list
fruits.sort() # sort list in ascending order
print(fruits)
```

```python
fruits.reverse() # reverse the list
print(fruits)
```

```python
## remove all item from list
fruits.clear()
print(fruits)
```

Slicing List


```python
num = [5, 2, 9, 1, 5, 6]
print(num[2:5])
print(num[:5])
print(num[3:]) # 1 5 6
print(num[::2]) # step size 2
print(num[::-1]) # reverse the list

```

```python
num[::3] #step size 3
```

```python
print(num)
num[::-2] # step size -2
```

iterate over list


```python
for number in num:
    print(number)   
```

```python
## iterating with index ----> enumerate()

for index,number in enumerate(num):
    print(f"Index: {index}, Number: {number}")  
   
```

```python
## List comprehensions
lst=[]
for x in range(10):
    lst.append(x**2)

    #print(f"{lst}")
print(lst)
```

 List comprehensions

 Basic syntax  [expression for item in iterable]

 with conditional logic [expression for item in iterable if condition
 ]

 Nested List Comprehesion [expression for item1 in iterable for item2 in iterable2]

```python
### Basic List Comprehension

square=[num**2 for num in range(10)]
print(square)
```

```python
## List comprehison with condition
lst=[]
for i in range(10):
    if i%2==0:
       lst.append(i)

print(lst)
```

```python
even_num=[i for i in range(10) if i%2==0]
print(even_num)
```

```python
## nested list comprehension

lst1=[1,2,3,4]
lst2=['a','b','c','d']

pair=[(i,j) for i in lst1 for j in lst2]

print(pair)
 
```

```python
words = ["hello", "world", "python", "programming"]
lengths = [len(word) for word in words] 
print(lengths)
```
