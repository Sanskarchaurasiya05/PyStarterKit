Dictionaries

Dictoneries are unordered collections of items. They store data in key-value pairs. Keys must be unique and immutable , while values can be of any type.

```python
## Creating Dictonaries
# here we use flower brases

empty_dict={} 
print(type(empty_dict))  
```

```python
empty_dict=dict()
print(type(empty_dict))
```

```python
student={"name":"Alice","age":25,"courses":["Math","CompSci"]}
print(student)
```

```python
## accessing Dictonary elements
print(student["name"])
print(student.get("age"))

## acessing using get() method
print(student.get("courses"))
print(student.get("phone","Not Found"))  # if key not found return default value
```

```python
# Modifying Dictionaries elements
student["age"]=26  # update age
student["phone"]="555-5555"  # add new key-value pair   
print(student)
```

```python
## deleting Dictionary elements
del student["courses"]  # delete key-value pair
print(student)
```

```python
## Dictionary Methods

Keys=student.keys()
print(Keys)
values=student.values()
print(values)

items=student.items()
print(items)
```

```python
# shallow copy
student2=student
print(student)
print(student2)
```

```python
student["name"]="krish"  # modify original dictionary
print(student)
print(student2)  # copied dictionary remains unchanged
```

```python
student_copy=student.copy()  # create a shallow copy using copy() method
print(student)
print(student_copy)
```

```python
student["name"]="Bob"  # modify original dictionary
print(student)
print(student_copy)  # deep copied dictionary remains unchanged
```

```python
## Iterating over Dictionaries
## you can use loops to iterate over keys, values, or key-value pairs in a dictionary.

# Iterating over keys
for key in student:
    print(key)
```

```python
#iterating over values
for value in student.values():
    print(value)
```

```python
#iterate over items
for key, value in student.items():
    print(f"{key}: {value}")
```

```python
## Nested Dictionaries
students={
    "student1":{"name":"Alice","age":25},
    "student2":{"name":"Bob","age":22},
}
print(students)
```

```python
## Access nested dictionary elements
print(students["student1"]["name"])  # Access name of student1
print(students["student1"]["age"])   # Access age of student2
```

```python
## iterating over nested dictionaries
for student_id, info in students.items():
    print(f"{student_id}:")
    for key, value in info.items():
        print(f"  {key}: {value}")
        
```
