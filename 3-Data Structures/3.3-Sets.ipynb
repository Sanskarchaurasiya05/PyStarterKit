Sets


```python
## How to create a Set

my_set={1,2,3,4,5}
print(my_set)
print(type(my_set))
```

```python
my_empty_set=set()
print(type(my_empty_set))
```

```python
# typecast to set

my_set=set([1,2,3,4])
print(my_set)
```

```python
my_empty_set=set([1,2,3,4,5,6])
print(my_empty_set)  

```

```python
# Basics Sets Operation
# Adding and Removing Elements in a Set

my_set.add(6)
my_set.add(7)
print(my_set)
```

```python
# remove the element from the set
my_set.remove(3)
print(my_set)
```

```python
my_set.remove(10)  # This will raise a KeyError because 10 is not in the set
```

```python
my_set.discard(10)  # This will not raise an error
```

```python
## pop method
popped_element=my_set.pop()
print(popped_element)
print(my_set)
```

```python
## clear all the elements

my_set.clear()
print(my_set)
```

```python
## Set Membership Test
my_set={1,2,3,4,5}
print(3 in my_set)  # True
print(10 in my_set) # False     
```

```python
## Mathematical Set Operations
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# Union
union_set=set1.union(set2)
print(union_set)

# Intersection
intersection_set=set1.intersection(set2)
print(intersection_set)

set1.intersection_update(set2) # Updates set1 to the intersection
print(set1)
```

```python
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

# Difference
difference_set=set1.difference(set2)
print(difference_set)

set1.difference_update(set2) # Updates set1 to the difference
print(set1)
```

```python
set2.difference(set1)  # Returns elements in set2 not in set1
```

```python
## Symetric differnece  ----give only unique elements from both sets
set1={1,2,3,4,5,6}
set2={4,5,6,7,8,9}

sym_diff_set=set1.symmetric_difference(set2)
print(sym_diff_set)
```

```python
## Sets Methosds

set1={1,2,3,4,5}
set2={4,5,6,7,8}

#is subset
print(set1.issubset(set2)) 

#is superset ---> wheather set2 contains all elements of set1
print(set2.issuperset(set1))
```

```python
# remove duplicates from a list using sets
lst=[1,2,2,3,4,4,5,5,5]

set(lst)
```

```python
## counting unique word in a text
text="This is a sample text with several words. This text is for counting unique words."
words=text.split()

# convert list of words to set to get unique words
unique_words=set(words)
print(unique_words)
print(len(unique_words))
```
