```python
## if statement : evaluates the condition inside the parentheses. If the condition is true, the block of code inside the if statement is executed.
age=20

if age>=20:
    print("You are eligible to vote.")
    
```

```python
##else statement : executes a block of code when the condition in the if statement is false.

age=17
if age>=18:
    print("You are eligible to vote.") 
else:
    print("You are not eligible to vote.") 
```

```python
##elif statement : allows you to check multiple conditions in sequence. If the first condition is false, it checks the next one, and so on.
age=20
if age<13:
    print("You are a child.")
elif age<20:
    print("You are a teenager.")
else:
    print("You are an adult.")
    
```

```python
##Nested conditional statements : you can place an if statement inside another if statement to create nested conditions.
#check a given number is even,odd,negative or zero
num=int(input("Enter a number: "))  
if num>=0:
    if num==0:
        print("The number is zero.")
    else:
        if num%2==0:
            print("The number is even.")
        else:
            print("The number is odd.")
else:
    print("The number is negative.")
```
