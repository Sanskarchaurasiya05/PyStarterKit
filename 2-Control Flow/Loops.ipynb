```python
## String

str="Krish Naik"

for i in str:
    print(i)
```

```python
## while Loop

count=0
while count<5:
    print(count)
    count+=1
```

```python
count=0

while count%2==0:
    print(count)
    count+=1
```

```python
## Loop Control Statements

## break Statement

for i in range(10):
    if i==5:
        break
    print(i)
```

```python
## continue Statement

for i in range(10):
    if i%2==0:
        continue
    print(i)
```

```python
## pass Statement
## the pass statement is null operation; nothing happens when it is executed. It is used as a placeholder.

for i in range(5):
    if i==3:
        print("Three")
        pass
    print(i)
```
