```python
##Arithmetic operations

a=10
b=3
res=a+b      # Addition
print(res)
res=a-b      # Subtraction
print(res)
res=a*b      # Multiplication
print(res)
res=a/b      # Division (float)
print(res)
res=a//b     # Division (floor)
print(res)
res=a%b      # Modulus
print(res)
res=a**b     # Exponentiation
print(res)
```

```python
## comparison operators
print(a==b)   # Equal to
print(a!=b)   # Not equal to
print(a>b)    # Greater than
print(a<b)    # Less than

print(a>=b)   # Greater than or equal to
print(a<=b)   # Less than or equal to
```
